## 高维数据可视化

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
y=iris.target
X=iris.data
print(X.shape)
import pandas as pd
pd.DataFrame(X)
```

```python
pca=PCA(n_components=2)
model_pca=pca.fit(X)
X_dr=model_pca.transform(X)
X_dr
```

```python
y
#三分类变量
```

```python
X_dr[y==0,0],X_dr[y==0,1]#取出样本为y==0,第一个特征
```

```python

plt.figure(dpi=400)
plt.scatter(X_dr[y==0,0],X_dr[y==0,1],color='red',label=iris.target_names[0])
plt.scatter(X_dr[y==1,0],X_dr[y==1,1],color='blue',label=iris.target_names[1])
plt.scatter(X_dr[y==2,0],X_dr[y==2,1],color='orange',label=iris.target_names[2])
plt.legend()

plt.title("PCA of iris data")
plt.show()

```

## 探索降维的数据


```python
model_pca.explained_variance_
```

```python
model_pca.explained_variance_ratio_
```

```python
model_pca.explained_variance_ratio_.sum()
```

```python
import numpy as np
pca_line=PCA().fit(X)
pca_line.transform(X)
plt.figure(dpi=400)
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_),label='property')
plt.xticks([1,2,3,4])
plt.xlabel("number of component after dimension reduction")
plt.ylabel("cumulative explained variance")
plt.legend()
plt.show()
```
