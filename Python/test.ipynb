# BP神经网络


```python
import pandas as pd
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from sklearn.metrics import r2_score

device=torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device
data=pd.read_excel("附件1BP.xlsx")
data_train=data.iloc[:,[1,2,3,4,5]]
data_target=data.iloc[:,[13]]
train=torch.tensor(data_train.values)#将pandas转torch
train=train.to(torch.float32)#将torch中的类型转化为float，因为有时pandas中格式不统一
train=train.to(device)
train.shape
train[:,0]/=250
train[:,1]/=0.05
train[:,2]/=250
train[:,3]/=3
train[:,4]-=200
train[:,4]/=250
train_1=train[0:90,:]
test=train[90:114,:]
target=torch.tensor(data_target.values).to(torch.float32).to(device)
target/=100
# data_train=data[]
class DNN(nn.Module):
    def __init__(self,n_feature,n_hidden_1,n_hidden_2,n_output) -> None:
        super(DNN,self).__init__()
        self.lc1=nn.Linear(n_feature,n_hidden_1)
        # self.lc2=nn.Linear(n_hidden_1,n_hidden_2)
        self.lc2=nn.Linear(n_hidden_1,n_output)
        self.lc3=nn.Linear(n_hidden_2,n_output)
        self.relu=nn.ReLU()

    def forward(self,x):
        x=self.relu(self.lc1(x))
        # x=self.relu(self.lc2(x))
        # x=self.lc3(x)
        x=self.lc2(x)
        return x

num=200000
num_feature=5
num_hidden_layer1=8
num_hidden_layer2=6
num_output=1

net=DNN(num_feature,num_hidden_layer1,num_hidden_layer2,num_output).to(device)

optimizer=torch.optim.SGD(net.parameters(),lr=0.001,momentum=0.05)
loss_fn=nn.MSELoss()

net.train()
for t in range(num):
    optimizer.zero_grad()
    prediction =net(train_1)
    loss = loss_fn(prediction,target[0:100,:])
    #将所有参数的梯度全部降为0，梯度值保留在这个里面
    loss.backward()    #反向传递过程
    optimizer.step() 
    #优化梯度
    if t%100==0:
          print(f"已训练{t}次，loss:{loss}")
```

```python
net.eval()
pred_1=net(test)
loss_fn(pred_1,target[100:114,:])
# pred_1-target[90:114,:]
```
