```python
import pandas as pd

data=pd.read_excel("附件2.xlsx",skiprows=2)
data.columns=['时间','转化率','乙烯选择性','C4烯烃选择性','乙醛选择性','脂肪醇选择性','甲基苯甲醛选择性','其他']
```

```python
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif']=['SimHei']

fig=plt.figure(figsize=(7,4),dpi=150)
ax1=fig.add_subplot(111)

x=data['时间']
y=data['转化率']

ax1.plot(x,y)
ax1.set_ylabel('乙醇转化率')
ax1.set_title("350摄氏度下乙醇转化率和各种生物的选择性随时间变化曲线")

ax2=ax1.twinx()
ax2.plot(x,data['乙烯选择性'])
ax2.plot(x,data['C4烯烃选择性'])
ax2.plot(x,data['乙醛选择性'])
ax2.plot(x,data['脂肪醇选择性'])
ax2.plot(x,data['甲基苯甲醛选择性'])
ax2.plot(x,data['其他'])

ax2.set_ylabel('选择性')
plt.legend(data.columns[1:],fontsize=8)
plt.show()
```
